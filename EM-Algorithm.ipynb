```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data)
Y = pd.DataFrame(iris.target)
X.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
Y.columns = ['Targets']
colorMaps = np.array(['red', 'lime', 'black'])
```

```python
kMeansModel = KMeans(n_clusters=3)
kMeansModel.fit(X)
```

```python
scaler = StandardScaler()
xsa = scaler.fit(X)
xsa = scaler.transform(X)
xs = pd.DataFrame(xsa, columns=X.columns)
gmm = GaussianMixture(n_components=3)
gmm.fit(xs)
yPredCluster = gmm.predict(xs)
```

```python
plt.figure(figsize=(14, 7))
plt.subplot(1, 3, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colorMaps[Y.Targets])
plt.xlabel('Petal_Length')
plt.ylabel('Petal_Width')
plt.title('Real Cluster')
plt.subplot(1, 3, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colorMaps[kMeansModel.labels_])
plt.xlabel('Petal_Length')
plt.ylabel('Petal_Width')
plt.title('KMeans Cluster')
plt.subplot(1, 3, 3)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colorMaps[yPredCluster])
plt.xlabel('Petal_Length')
plt.ylabel('Petal_Width')
plt.title('GMM Cluster')
```
